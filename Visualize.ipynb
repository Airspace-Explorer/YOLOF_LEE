```python
import json

def getInformationsFromLogFile(filePath):

    with open(filePath) as file:
        for line in file:
            oneJsonData = json.loads(line)
            if oneJsonData["mode"] == 'train':
                train_learning_rate_list.append(oneJsonData["lr"])  # Learning Rate
                train_loss_cls_list.append(oneJsonData["loss_cls"])  # Classification Loss
                train_loss_bbox_list.append(oneJsonData["loss_bbox"])  # Bbox Regression Loss
                train_loss_list.append(oneJsonData["loss"])  # Total Loss
            elif oneJsonData["mode"] == 'val':
                val_learning_rate_list.append(oneJsonData['lr'])  # Learning Rate
                val_ap50_list.append(oneJsonData['AP50'])  # AP50 (IOU50 적용시 Average Precision)
                val_map_list.append(oneJsonData['mAP'])  # mAP

    return (
        train_learning_rate_list,
        train_loss_cls_list,
        train_loss_bbox_list,
        train_loss_list,
        val_learning_rate_list,
        val_ap50_list,
        val_map_list
    )
```

```python
getInformationsFromLogFile("./data/None.log.json")
```

```python
import matplotlib.pyplot as plt

# Assuming train_learning_rate_list has more than 5000 elements
x_values = range(5000)
train_learning_rate_subset = train_learning_rate_list[:5000]

fig = plt.figure(figsize=(4, 4), facecolor='linen')

ax1 = fig.add_subplot(1, 1, 1)  # 행 1개, 열 1개 중 첫번째 오브젝트 선택
ax1.plot(x_values, train_learning_rate_subset)
ax1.set_title("Learning rate trend during model training", fontsize=15)
ax1.set_xlabel("example", fontsize=10)
ax1.set_ylabel("train_learning_rate", fontsize=10)

plt.show()
```
